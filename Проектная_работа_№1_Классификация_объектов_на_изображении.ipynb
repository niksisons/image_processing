### **Базовая реализация:**

1. **Сбор изображения:**

2. **Предварительная обработка изображения:**
   - Преобразуйте изображение в оттенки серого.

3. **Обнаружение кругов (монет):**

  - Для отображения текста воспользуйтесь `cv2.putText()`.
- **Обработка ошибок:**
  - Обрабатывайте случаи, когда преобразование Хафа не обнаруживает ни одной монеты.
  - Предусмотрите проверку входных данных и вывод информативных сообщений для пользователя.



```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

def classify_coin_by_radius(radius):
    if 100 <= radius < 130:
        return 1
    elif 130 <= radius < 153:
        return 2
    elif 153 <= radius < 201:
        return 5
    else:
        return None

def count_lines_in_coin(gray_full, x, y, r):
    size = 2 * r
    x1 = max(0, x - r)
    y1 = max(0, y - r)
    x2 = min(gray_full.shape[1], x + r)
    y2 = min(gray_full.shape[0], y + r)

    roi = gray_full[y1:y2, x1:x2]
    if roi.size == 0:
        return 0

    blurred_roi = cv2.GaussianBlur(roi, (5, 5), 1)
    edges_roi = cv2.Canny(blurred_roi, 50, 150)

    lines = cv2.HoughLinesP(
        edges_roi,
        rho=1,
        theta=np.pi / 180,
        threshold=20,          # Порог голосов
        minLineLength=15,      # Мин. длина линии (в пикселях)
        maxLineGap=5           # Макс. разрыв между сегментами
    )

    line_count = 0
    if lines is not None:
        line_count = len(lines)

    return line_count, edges_roi

def main(image_path):
    img = cv2.imread(image_path)
    if img is None:
        print(f" Ошибка: не удалось загрузить изображение '{image_path}'")
        return 0

    cv2_imshow(img)

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    print("→ Серое изображение:")
    cv2_imshow(gray)

    blurred = cv2.GaussianBlur(gray, (9, 9), 2)
    print("→ Размытое изображение:")
    cv2_imshow(blurred)

    edges = cv2.Canny(blurred, 50, 150)
    print("→ Края (Canny):")
    cv2_imshow(edges)

    circles = cv2.HoughCircles(
        blurred,
        cv2.HOUGH_GRADIENT,
        dp=1,
        minDist=80,
        param1=50,
        param2=150,
        minRadius=110,
        maxRadius=200
    )

    output_img = img.copy()
    total_sum = 0
    coin_count = 0

    if circles is not None:
        circles = np.uint16(np.around(circles))
        print("Обнаруженные монеты:")
        print("№\tРадиус\tДиаметр\tНоминал\tЛиний")
        print("-" * 50)

        for i, (x, y, r) in enumerate(circles[0, :], 1):
            denomination = classify_coin_by_radius(r)
            label = f"{denomination} rub" if denomination else "?"

            line_count, edges_roi = count_lines_in_coin(gray, x, y, r)

            if denomination:
                total_sum += denomination
                coin_count += 1

            print(f"{i}\t{r}\t{2*r}\t{label}\t{line_count}")

            cv2.circle(output_img, (x, y), r, (0, 255, 0), 2)
            cv2.circle(output_img, (x, y), 2, (0, 0, 255), 3)
            cv2.putText(output_img, label, (x - r, y - r - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)

            if line_count > 0 and i <= 3:
                roi_debug = cv2.cvtColor(edges_roi, cv2.COLOR_GRAY2BGR)
                lines = cv2.HoughLinesP(
                    edges_roi, 1, np.pi/180, threshold=20,
                    minLineLength=15, maxLineGap=5
                )
                if lines is not None:
                    for line in lines:
                        x1l, y1l, x2l, y2l = line[0]
                        cv2.line(roi_debug, (x1l, y1l), (x2l, y2l), (0, 255, 0), 1)
                print(f"→ ROI монеты #{i} (линии: {line_count}):")
                cv2_imshow(roi_debug)

        print(f"\n Всего монет: {coin_count}")
        print(f" Общая сумма: {total_sum} рублей")
    else:
        print(" Монеты не обнаружены.")
        total_sum = 0

    print("\n Результат:")
    cv2_imshow(output_img)
    cv2.imwrite("result_with_hough_lines.jpg", output_img)
    print("Сохранено как 'result_with_hough_lines.jpg'")

    return total_sum


IMAGE_PATH = "scale_1200.jpg"
main(IMAGE_PATH)
```
