### **Базовая реализация:**

2. **Предварительная обработка изображения:**

- **Отображение результатов:**

- **Обработка ошибок:**

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

def classify_coin_by_radius(radius):
    if 100 <= radius < 130:
        return 1
    elif 130 <= radius < 153:
        return 2
    elif 153 <= radius < 201:
        return 5
    else:
        return None


def count_lines_in_coin(gray_full, x, y, r):
    size = 2 * r
    x1 = max(0, x - r)
    y1 = max(0, y - r)
    x2 = min(gray_full.shape[1], x + r)
    y2 = min(gray_full.shape[0], y + r)

    roi = gray_full[y1:y2, x1:x2]
    if roi.size == 0:
        return 0, None

    blurred_roi = cv2.GaussianBlur(roi, (5, 5), 1)
    edges_roi = cv2.Canny(blurred_roi, 50, 150)

    lines = cv2.HoughLinesP(
        edges_roi,
        rho=1,
        theta=np.pi / 180,
        threshold=20,
        minLineLength=15,
        maxLineGap=5
    )

    line_count = 0
    if lines is not None:
        line_count = len(lines)

    return line_count, edges_roi


def count_grid_lines_in_coin(gray_full, x, y, r, spacing=4):
    """
    Вторая реализация: создаём маску круга (монеты) и поверх неё рисуем равномерно
    расположенные горизонтальные линии с шагом spacing пикселей. Считаем, сколько
    таких линий пересекают область круга.

    Возвращает: (count, roi_debug)
    """
    x1 = max(0, x - r)
    y1 = max(0, y - r)
    x2 = min(gray_full.shape[1], x + r)
    y2 = min(gray_full.shape[0], y + r)

    roi = gray_full[y1:y2, x1:x2]
    if roi.size == 0:
        return 0, None

    h, w = roi.shape
    # Создаём маску круга в ROI
    mask = np.zeros((h, w), dtype=np.uint8)
    center = (w // 2, h // 2)
    rr = min(center[0], center[1], w - center[0], h - center[1])
    cv2.circle(mask, center, rr, 255, -1)

    # Подготовка отладочного цветного ROI (для рисования линий)
    roi_color = cv2.cvtColor(roi, cv2.COLOR_GRAY2BGR)

    count = 0
    # Перебираем горизонтальные линии внутри ROI с заданным шагом
    for yy in range(0, h, spacing):
        # Проверяем есть ли ненулевые пиксели маски на этой строке
        if np.any(mask[yy, :] > 0):
            count += 1
            # Рисуем линию для отладки
            cv2.line(roi_color, (0, yy), (w - 1, yy), (0, 255, 0), 1)

    # Нарисуем контур круга на отладочном ROI
    cv2.circle(roi_color, center, rr, (0, 0, 255), 1)

    return count, roi_color


def classify_by_grid_count(line_count, thresholds=None):
    """
    Классификация по количеству линий, помещающихся в кругу.
    При spacing=4:
    - монета 1 рубль (маленькая) ~ 55-65 линий
    - монета 2 рубля (средняя) ~ 65-75 линий
    - монета 5 рублей (большая) ~ 75-85 линий
    """
    if line_count == 0:
        return None

    # При spacing=4 используем следующие пороги
    if line_count < 65:  # меньше 65 линий - это 1 рубль
        return 1
    elif line_count < 75:  # от 65 до 75 линий - это 2 рубля
        return 2
    else:  # больше 75 линий - это 5 рублей
        return 5


def main(image_path):
    img = cv2.imread(image_path)
    if img is None:
        print(f" Ошибка: не удалось загрузить изображение '{image_path}'")
        return 0

    cv2_imshow(img)

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    print("→ Серое изображение:")
    cv2_imshow(gray)

    blurred = cv2.GaussianBlur(gray, (9, 9), 2)
    print("→ Размытое изображение:")
    cv2_imshow(blurred)

    edges = cv2.Canny(blurred, 50, 150)
    print("→ Края (Canny):")
    cv2_imshow(edges)

    circles = cv2.HoughCircles(
        blurred,
        cv2.HOUGH_GRADIENT,
        dp=1,
        minDist=80,
        param1=50,
        param2=150,
        minRadius=110,
        maxRadius=200
    )

    output_img = img.copy()
    total_sum = 0
    total_sum_grid = 0
    coin_count = 0

    if circles is not None:
        circles = np.uint16(np.around(circles))
        print("Обнаруженные монеты:")
        print("№\tРадиус\tДиаметр\tНоминал(радиус)\tЛиний(сетка)\tНоминал(сетка)")
        print("-" * 80)

        for i, (x, y, r) in enumerate(circles[0, :], 1):
            denomination = classify_coin_by_radius(r)

            # Используем spacing=4
            grid_line_count, roi_grid_debug = count_grid_lines_in_coin(gray, x, y, r, spacing=4)
            denomination_grid = classify_by_grid_count(grid_line_count)

            label = f"{denomination}" if denomination else "?"
            label_grid = f"{denomination_grid}" if denomination_grid else "?"

            if denomination:
                total_sum += denomination
                coin_count += 1
            if denomination_grid:
                total_sum_grid += denomination_grid

            print(f"{i}\t{r}\t{2*r}\t{label}\t\t{grid_line_count}\t\t{label_grid}")

            cv2.circle(output_img, (x, y), r, (0, 255, 0), 2)
            cv2.circle(output_img, (x, y), 2, (0, 0, 255), 3)

            text = f"{label_grid}rub"
            fontScale = r / 70
            thickness = max(1, int(r / 50))
            font = cv2.FONT_HERSHEY_COMPLEX
            (text_width, text_height), baseline = cv2.getTextSize(text, font, fontScale, thickness)
            text_x = x - text_width // 2
            text_y = y + text_height // 2

            cv2.putText(output_img, text, (text_x, text_y),
                       font, fontScale, (255, 255, 255), thickness + 2)
            cv2.putText(output_img, text, (text_x, text_y),
                       font, fontScale, (0, 0, 255), thickness)

            if roi_grid_debug is not None:
                print(f"→ ROI монеты #{i} — сетка (lines: {grid_line_count}):")
                cv2_imshow(roi_grid_debug)

        print(f"\n Всего монет: {coin_count}")
        print(f" Общая сумма (по радиусу): {total_sum} рублей")
        print(f" Общая сумма (по сетке): {total_sum_grid} рублей")
    else:
        print(" Монеты не обнаружены.")
        total_sum = 0

    print("\n Результат:")
    cv2_imshow(output_img)
    cv2.imwrite("result_with_grid.jpg", output_img)
    print("Сохранено как 'result_with_grid.jpg'")

    return total_sum


IMAGE_PATH = "scale_1200.jpg"
main(IMAGE_PATH)
```
